The case of $A$ is a square matrix.

$A = QR$, where $Q$ is an orthogonal matrix and $R$ is a upper triangular matrix.

```python
import numpy as np
import numpy.random as rn
import numpy.linalg as la
```

```python
#QR for square matrix
def SQR(A,tol=10**(-14)):
    m,n = A.shape
    Q = np.mat(np.zeros((m,n)))
    R = np.mat(np.zeros((n,n)))
    for i in range(n):
        Q[:,i] = A[:,i].copy()
        for j in range(i):
            R[j,i] = float(Q[:,j].T*Q[:,i])
            Q[:,i] = Q[:,i] - R[j,i]*Q[:,j]
        if abs(la.norm(Q[:,i]))>tol:
            R[i,i] = la.norm(Q[:,i])
            Q[:,i] = Q[:,i]/R[i,i]
       
    return Q, R
```

```python
A = np.mat(rn.randn(3,3))
```

```python
Q,R = SQR(A)
```

QR for general matrix

```python
def QR(A,tol=10**(-14)):
    m,n = A.shape
    if m>=n:
        Q = np.mat(np.zeros((m,n)))
        R = np.mat(np.zeros((n,n)))
    else:
        Q = np.mat(np.zeros((m,m)))
        R = np.mat(np.zeros((m,n)))
    flag = 0
    i = 0
    while i<m:
        Q[:,i] = A[:,i].copy()
        for j in range(i):
            R[j,i] = float(Q[:,j].T*Q[:,i])
            Q[:,i] = Q[:,i] - R[j,i]*Q[:,j]
        if abs(la.norm(Q[:,i]))>tol:
            R[i,i] = la.norm(Q[:,i])
            Q[:,i] = Q[:,i]/R[i,i]
            if flag==1:
                Q[:,[i-1,i]] = Q[:,[i,i-1]]
                R[[i-1,i],:] = R[[i-1,i],:]
                i-=1
        else:
            flag = 1
        i+=1
    R[:,m:] = Q.T*A[:,m:]
       
    return Q, R
```

```python
A = np.mat(rn.randn(30,100))
Q,R = QR(A)
```
