Find $P_n(x)$ such that $P_n(x_i)=y_i$ for $i=0,1,...,n$

```python
import numpy as np
import numpy.random as rn
from scipy.interpolate import interp1d
from scipy.interpolate import lagrange
```

```python
x = np.linspace(-2*np.pi,np.pi,200)
y = 2*np.sin(x) - 3*np.cos(x)
```
